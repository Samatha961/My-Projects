```python
# Task Manager System
class TaskManager:
    def __init__(self):
        self.tasks = []  # List to store tasks

    def add_task(self, task_name, due_date):
        """Add a new task with a name and due date."""
        task = (task_name, due_date, False)  # Tuple (task_name, due_date, completed)
        self.tasks.append(task)
        print(f"Task '{task_name}' added with due date {due_date}.")

    def complete_task(self, task_index):
        """Mark a task as completed."""
        if 0 <= task_index < len(self.tasks):
            task_name, due_date, _ = self.tasks[task_index]
            self.tasks[task_index] = (task_name, due_date, True)
            print(f"Task '{task_name}' marked as completed.")
        else:
            print("Invalid task index.")

    def remove_task(self, task_index):
        """Remove a task by index."""
        if 0 <= task_index < len(self.tasks):
            task_name, _, _ = self.tasks.pop(task_index)
            print(f"Task '{task_name}' removed.")
        else:
            print("Invalid task index.")

    def display_tasks(self):
        """Display all tasks categorized by status."""
        if not self.tasks:
            print("No tasks available.")
            return
        
        print("\nTasks:")
        for index, (task_name, due_date, completed) in enumerate(self.tasks):
            status = "Completed" if completed else "Incomplete"
            print(f"{index}. {task_name} (Due: {due_date}) - {status}")

# Example usage of the TaskManager
if __name__ == "__main__":
    manager = TaskManager()
    
    # Adding tasks
    manager.add_task("Finish homework", "2024-09-30")
    manager.add_task("Clean the house", "2024-09-28")
    manager.add_task("Buy groceries", "2024-09-27")
    
    # Displaying tasks
    manager.display_tasks()
    
    # Completing a task
    manager.complete_task(1)  # Complete "Clean the house"
    
    # Displaying tasks again
    manager.display_tasks()
    
    # Removing a task
    manager.remove_task(0)  # Remove "Finish homework"
    
    # Displaying tasks again
    manager.display_tasks()
```
